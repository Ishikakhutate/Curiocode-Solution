# Bivariant Analysis

```python
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival count by Gender')
plt.show()
```

```python
selected_features = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']

# Creating a pairplot
sns.pairplot(data[selected_features], hue='Survived', kind='count', palette='husl')
plt.show()
```

```python
selected_features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

# Plotting boxplots for each feature
for feature in selected_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Survived', y=feature, data=data)
    plt.title(f'Boxplot of {feature} by Survival Status')
    plt.xlabel('Survived')
    plt.ylabel(feature)
    plt.show()
```

```python
sns.pairplot(data=data)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Flatten the axes array
axes = axes.flatten()

# List of categorical variables
categorical_vars = ['Survived', 'Sex', 'Pclass', 'SibSp', 'Parch']

# Loop through each categorical variable and create count plots
for i, var in enumerate(categorical_vars):
    sns.countplot(data=data, x=var, ax=axes[i])
    axes[i].set_title(f'Count Plot of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Flatten the axes array
axes = axes.flatten()

# List of numerical variables
numerical_vars = ['Age', 'Fare','Survived','Pclass', 'SibSp', 'Parch']

# Loop through each numerical variable and create box plots
for i, var in enumerate(numerical_vars):
    sns.boxplot(data=data[var], ax=axes[i])
    axes[i].set_title(f'Boxplot of {var}')
    axes[i].set_xlabel(var)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived')
plt.title('Scatterplot of Age vs Fare with Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()
```

# Model Creation  
(Logistic Regression)

```python
X = data.drop(['Survived'],axis=1)
y = data.Survived
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 20)
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#Using Logistic Regression
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
y_pred
```

# Evalution Model

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)        
```

```python
recall=recall_score(y_test,y_pred)
recall
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```
